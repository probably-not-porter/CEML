# Plot Final Epoch
Plot the final epoch of each results bundle (output folder) in the given directory.

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import os

def create_plot(directory):
    groups = []
    val1 = [] # training loss
    val2 = [] # valid loss
    val3 = [] # error rate
    results = {
        "Training Loss": [],
        "Valid Loss": [],
        "Error Rate": []
    }

    for dir in os.listdir(directory):
        f = directory + "/" + dir + "/epoch_stats.csv"

        with open(f) as csv_file:
            final_line = csv_file.readlines()[-1].split(",")
            groups.append(dir)
            val1.append(float(final_line[1]))
            val2.append(float(final_line[2]))
            val3.append(float(final_line[3]))

    fig, ax = plt.subplots()

    # Stacked bar chart
    ax.bar(groups, val1, label="training Loss")
    ax.bar(groups, val2, bottom = val1, label="Valid Loss")
    ax.bar(groups, val3, bottom = np.add(val1, val2), label="Error Rate")
    ax.legend()
    plt.title("Final training results (last epoch)")
    plt.xticks(rotation=45, ha='right')
    plt.show() 
```

```python
create_plot('Apr23_2023')
```

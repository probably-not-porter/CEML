## STEP 1
Import the datastructure (.py file), this will give the user access to the same settings and operations as my previous demo, but in a much more abstract manner.

```python
# IMPORTS
from dataglob import DataGlob
import os
```

## STEP 2
Create a dataglob. This will require a path for an input, which is a directory of images for the AI to work with, and out output directory, which is a location where the `DataGlob` can store all of its assets and temporary work.

```python
in_dir = os.getcwd() + "/../training/plant-id"
out_dir = os.getcwd()

# create glob
PlantGlob = DataGlob(in_dir, out_dir)
```

## STEP 3
Now that we have a `DataGlob`, we can start turning some knobs. For this example, we will use a combination of the Prewitt edge detector, Canny Edge Detector, and the original images. These three options will be compressed as the RGB channels of a single JPEG that the AI can work with

```python
# adjust some knobs
PlantGlob._overwrite = True
PlantGlob.set_configuration("prewitt", True)
PlantGlob.set_configuration("canny_auto", True)
PlantGlob.set_configuration("original", True)

# see the knobs
PlantGlob.show_configuration()
```

## STEP 4
This step will create a database in the output folder of the `DataGlob` by processing all of the images in each label, and putting them in a parallel folder in the output directory.

```python
PlantGlob.prepare_database()
```

## STEP 5
This step will convert the images in the output directory into a `databunch (from FastAI, NOT a DataGlob)`, which is an object that FastAI can work with to train a model.

```python
PlantGlob.create_databunch()
PlantGlob.show_databunch()
```

## STEP 6
This step is the final method that we will use from the `DataGlob` at this point. It will use what has been generated in the output directory to create a model using a Resnet34 architechture.

```python
PlantGlob.create_model()
```
